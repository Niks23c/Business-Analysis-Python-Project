##🧑‍🦳🧑‍🦲 Churn by age Group

```python
df_sc_pivot = (df.groupby('SeniorCitizen')["Churn"].value_counts().reset_index()).pivot(index='SeniorCitizen', columns='Churn', values='count')
df_sc_pivot
df_sc_percent = df_sc_pivot.div(df_sc_pivot.sum(axis=1),axis=0)

import matplotlib.pyplot as plt

df_sc_percent.plot(kind='bar', stacked=True, figsize=(8,5), colormap='viridis')
plt.title('Churn by citizen age')
plt.ylabel('Customer Proportion')
plt.xlabel('Citize Group')
plt.xticks(ticks=[0, 1], labels=['Non Senior citizen', 'Senior Citizen'])
plt.xticks(rotation=45)
plt.legend(title='Churn')
plt.show()
```

```python
df['tenure'].quantile([0.25, 0.50, 0.75])
```

## 🤵Churn by Citizen types

```python
import plotly.express as px

# categorizing the tebure

bins = [0, 12, 36, df['tenure'].max()]

labels = ['0-12', '13-36', '37+']

df['tenure_group'] = pd.cut(df['tenure'], bins=bins, labels=labels, include_lowest=True)

fig = px.pie(df, names='tenure_group', title='Customer Distribution by Tenure Group')

fig.update_layout(width=408, height=408)

fig.show()
```

```python
df.head()
```

## 🔎Reviewing monthly charges via tenure group

```python
Grouped = df.groupby(["tenure_group"],observed=True)["MonthlyCharges"].mean().reset_index()
fig = px.bar(Grouped, x='tenure_group', y='MonthlyCharges', text='MonthlyCharges',
title = "Avg Monthly Charges by Tenure Group" )
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside', textfont=dict(color='white', family='Arial white', size=14))
fig.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
    font=dict(color='white')
)
fig.update_layout(hoverlabel=dict(bgcolor='white',font_size=14, font_family='Bold', font_color='blue'))
fig.show()
```
